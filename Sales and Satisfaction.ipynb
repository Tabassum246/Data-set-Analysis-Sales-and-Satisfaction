## ***1.Label Encoding***

```python
import pandas as pd

# Example dataset
LE = pd.DataFrame(
    {
    "Name": ["A", "B", "C"],
    "Gender": ["Male", "Female", "Other"]
    })

# Apply Label Encoding
LE = LabelEncoder()
df["Gender_Encoded"] = LE.fit_transform(df["Gender"])

print(df)
```

## ***One-Hot Encoding***

```python
# Example dataset
df = pd.DataFrame({
    "Item": [1, 2, 3],
    "Color": ["Red", "Blue", "Green"]
})

# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=["Color"])
bool_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[bool_cols] = df_encoded[bool_cols].astype(int)

print(df_encoded)
```

## ***Ordinal Encoding***

```python
# Example dataset
df = pd.DataFrame({
    "Customer": ["A", "B", "C", "D"],
    "Satisfaction": ["Poor", "Average", "Good", "Excellent"]
})

# Define order of categories
categories = [["Poor", "Average", "Good", "Excellent"]]

# Apply Ordinal Encoding
oe = OrdinalEncoder(categories=categories)
df["Satisfaction_Encoded"] = oe.fit_transform(df[["Satisfaction"]])

print(df)
```

## ***Frequency (Count) Encoding***

```python
# Example dataset
df = pd.DataFrame({
    "City": ["Dhaka", "Chittagong", "Sylhet", "Dhaka", "Dhaka", "Chittagong"]
})

# Frequency encoding
freq_encoding = df["City"].value_counts()
df["City_Encoded"] = df["City"].map(freq_encoding)

df_unique = df.drop_duplicates(subset="City")[["City", "City_Encoded"]]
print(df_unique.reset_index(drop=True))


```
